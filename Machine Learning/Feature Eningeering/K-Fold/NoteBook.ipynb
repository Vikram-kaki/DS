```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(file2.iloc[:,:-1], file2.iloc[:,-1], train_size=0.7)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor


regressor = RandomForestRegressor(n_estimators=100)
```

```python
regressor.fit(X_train, y_train)
```

```python
vals = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error


print(mean_absolute_error(y_test, vals))
print(mean_squared_error(y_test, vals))
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],  
    'max_depth': [None, 10, 20],     
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]  
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model = GridSearchCV(RandomForestRegressor(), param_grid=param_grid, cv = 10, n_jobs= -1)
```

```python
model.fit(X_train, y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
estimator = model.best_estimator_
```

```python
estimator.predict(X_test)
```

```python
mean_squared_error(y_test, estimator.predict(X_test))
```

## K-FOLD

```python
from sklearn.model_selection import KFold

kFold = KFold(n_splits=10)

for test, train in kFold.split(file2):
    print("Train - ", train)
    print("Test -",test)
```

```python
X = file2.iloc[:,:-1]
y = file2.iloc[:,-1]
```

```python
erros = []


for train, test in kFold.split(file2):
    
    X_train, X_test = file2.loc[train], file2.loc[test]
    y_train, y_test = file2.loc[train], file2.loc[test]
    
    regressor = RandomForestRegressor(n_estimators=100)
    regressor.fit(X_train, y_train)
    
    erros.append(mean_squared_error(y_test, regressor.predict(X_test)))
    
```

```python
print(erros[erros.index(min(erros))])
```
